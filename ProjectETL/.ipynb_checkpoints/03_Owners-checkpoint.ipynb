```python
# Dependencies
import pandas as pd
import sqlalchemy

# Store filepath in a variable
owner_file = "Real_acct_owner/owners.csv"

# Read our Data file with the pandas library
owner_file_df = pd.read_csv(owner_file, encoding="ISO-8859-1")

owner_file_df['Account'] = owner_file_df['Account'].astype('int64')

owner_file_df = owner_file_df[["Account", "Name"]]
owner_file_df
```

```python
##Creating output file with only two columns
owner_file_df.to_csv("Real_acct_owner/owner_file_df_mysql.csv", index=False, header=True)
```

```python
##Reading newly create csv file with certain columns
owner_file ="Real_acct_owner/owner_file_df_mysql.csv"
owner_file = pd.read_csv(owner_file, encoding="ISO-8859-1", low_memory=False)


# Read data from newly created csv file.
owner_file[["Account", "Name"]]
```

```python
## Removing duplicate 

owner_file.drop_duplicates(subset='Account', inplace=True)
owner_file.to_csv("Real_acct_owner/owner_file_without_dup.csv", index=False, header=True)
```

```python
## connection for Mysql
engine = sqlalchemy.create_engine('mysql+pymysql://root:root@localhost:3306/hcad')

## reading file to insert into mysql
owner_data_ins =  pd.read_csv('Real_acct_owner/owner_file_without_dup.csv', encoding='latin-1',usecols=['Account', 'Name'])

##inserting data into mysql
owner_data_ins.to_sql('owner', con=engine, if_exists='append',index=False,chunksize=1)
```
