```python
# Dependencies
import pandas as pd
import sqlalchemy 

# Store filepath in a variable
acct_one = "Real_acct_owner/real_acct.csv"

# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
acct_one_df = pd.read_csv(acct_one, encoding="ISO-8859-1", low_memory=False)

acct_one_df['Account'] = acct_one_df['Account'].astype('int64')

## storing all data from csv file to dataframe

read_acct_mysql_df = acct_one_df[["Account", "TaxYear","Mailto","MailAddr","City","State","ZipCode","Neighbourhood","Total_land","LandValue","Assessed_value"]]
read_acct_mysql_df
```

```python
##Creating output file with only certain columns
read_acct_mysql_df.to_csv("Real_acct_owner/read_acct_mysql.csv", index=False, header=True)
```

```python
##Read newly create csv file with certain columns
acct_mysql ="Real_acct_owner/read_acct_mysql.csv"
read_acct_mysql_df = pd.read_csv(acct_mysql, encoding="ISO-8859-1", low_memory=False)


# Read data from newly created csv file.
read_acct_mysql_df[["Account", "TaxYear","Mailto","MailAddr","City","State","ZipCode","Neighbourhood","Total_land","LandValue","Assessed_value"]]
```

```python
## Removing duplicate 

read_acct_mysql_df.drop_duplicates(subset='Account', inplace=True)
read_acct_mysql_df.to_csv("Real_acct_owner/read_acct_without_dup.csv", index=False, header=True)
```
