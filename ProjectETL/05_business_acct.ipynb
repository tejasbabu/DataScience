```python
# Dependencies
import pandas as pd
import sqlalchemy

# Store filepath in a variable
business_file = "Real_acct_owner/t_business_acct.csv"

```

```python
# Read our Data file with the pandas library
business_file_df = pd.read_csv(business_file, encoding="ISO-8859-1")

business_file_df.drop_duplicates(subset='ACCOUNT', inplace=True)
business_file_df.dropna()

business_file_df['ACCOUNT'] = business_file_df['ACCOUNT'].astype('int64')
business_file_df.dtypes
```

```python
business_file_df = business_file_df[["ACCOUNT", "TAX_YEAR","O_NAME","OWNER","SITE_ADDRESS","SITE_CITY","SITE_STATE","APPRAISED_VALUE","PRIOR_APPRAISED_VALUE"]]
business_file_df 
```

```python
## Removing duplicate 

business_file_df.drop_duplicates(subset='ACCOUNT', inplace=True)
business_file_df.to_csv("Real_acct_owner/t_business_acct_dup.csv", index=False, header=True)
```

```python
## connection for Mysql
engine = sqlalchemy.create_engine('mysql+pymysql://root:root@localhost:3306/hcad')

## reading file to insert into mysql
business_data_ins =  pd.read_csv('Real_acct_owner/t_business_acct_dup.csv', encoding='latin-1',usecols=['ACCOUNT','TAX_YEAR','O_NAME','OWNER','SITE_ADDRESS','SITE_CITY','SITE_STATE','APPRAISED_VALUE','PRIOR_APPRAISED_VALUE'])

##inserting data into mysql
business_data_ins.to_sql('business_acct', con=engine, if_exists='append',index=False,chunksize=1)
```
