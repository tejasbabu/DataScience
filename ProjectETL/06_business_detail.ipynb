```python
# Dependencies
import pandas as pd
import sqlalchemy 

# Store filepath in a variable
business_detail = "Real_acct_owner/t_business_detail.csv"

# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
business_detail_df = pd.read_csv(business_detail, encoding="ISO-8859-1", low_memory=False)

business_detail_df['Account'] = business_detail_df['Account'].astype('int64')

## storing all data from csv file to dataframe

business_detail_df = business_detail_df[["Account", "Dept_code","Dept_code_de"]]
business_detail_df
```

```python
## Removing duplicate 

business_detail_df.drop_duplicates(subset='Account', inplace=True)
business_detail_df.to_csv("Real_acct_owner/t_business_detail_dup.csv", index=False, header=True)
```

```python
## connection for Mysql
engine = sqlalchemy.create_engine('mysql+pymysql://root:root@localhost:3306/hcad')

## reading file to insert into mysql
business_detail_data_ins =  pd.read_csv('Real_acct_owner/t_business_detail_dup.csv', encoding='latin-1',usecols=['Account', 'Dept_code','Dept_code_de'])

##inserting data into mysql
business_detail_data_ins.to_sql('business_detail', con=engine, if_exists='append',index=False,chunksize=1)
```
