```python
import pandas as pd
import sqlalchemy
import cx_Oracle
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy
```

```python
#defining mysql connection

engine = sqlalchemy.create_engine('mysql+pymysql://root:root@localhost:3306/hcad')
```

```python
value_city_df  = pd.read_sql('select * from investment;',engine)
value_city_df =value_city_df.set_index('city')
value_city_df 
```

```python
value_city_df.plot(kind="bar", figsize=(8,4), fontsize=10)
# Set a title for the chart
plt.title("Investment From Other City of US in Houston (in 100000)" ,  fontsize=10)
plt.show()
plt.tight_layout()
```

```python
value_zipcode_df  = pd.read_sql('select * from assesssedvalue;',engine)
value_zipcode_df =value_zipcode_df.set_index('zipcode')
value_zipcode_df 
```

```python
value_zipcode_df.plot(kind="bar", figsize=(8,4), fontsize=10)
# Set a title for the chart
plt.title("List of Zip code with highest Assessed value in Houston (in 100000000)" ,  fontsize=10)
plt.show()
plt.tight_layout()
```

```python
value_appraised_df  = pd.read_sql('select * from appraisedvalue;',engine)
value_appraised_df =value_appraised_df.set_index('city')
value_appraised_df 
```

```python
value_appraised_df.plot(kind="bar", figsize=(8,4), fontsize=10)
# Set a title for the chart
plt.title("List of City  with highest Appraised value in HCAD except Houston (in 100000)" ,  fontsize=10)
plt.show()
plt.tight_layout()
```
