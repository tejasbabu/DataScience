```python
# Dependencies
import pandas as pd
import sqlalchemy

# Store filepath in a variable
deeds_file = "Real_acct_owner/deeds.csv"

# Read our Data file with the pandas library
deeds_file_df = pd.read_csv(deeds_file, encoding="ISO-8859-1")

deeds_file_df
```

```python
## droping rown with null Account value
deeds_file_clean_df = deeds_file_df.dropna(subset = ['Account'])

##Creating output file without Null account number
deeds_file_clean_df.to_csv("Real_acct_owner/deeds_file_df.csv", index=False, header=True)
```

```python
# Store filepath in a variable
deeds_file_clean_df = "Real_acct_owner/deeds_file_df.csv"

# Read our Data file with the pandas library
deeds_file_clean_df = pd.read_csv(deeds_file_clean_df, encoding="ISO-8859-1")

deeds_file_clean_df['Account'] = deeds_file_clean_df['Account'].astype('int64')

deeds_file_clean_df = deeds_file_clean_df[["Account", "DateOfSale","Clerk_id","deed_id"]]
deeds_file_clean_df
```

```python
## Removing duplicate 

deeds_file_clean_df.drop_duplicates(subset='Account', inplace=True)
deeds_file_clean_df.to_csv("Real_acct_owner/deeds_file_clean_dup.csv", index=False, header=True)
```

```python
## connection for Mysql
engine = sqlalchemy.create_engine('mysql+pymysql://root:root@localhost:3306/hcad')

## reading file to insert into mysql
deeds_file_clean_ins =  pd.read_csv('Real_acct_owner/deeds_file_clean_dup.csv', encoding='latin-1',usecols=['Account', 'DateOfSale','Clerk_id','deed_id'],parse_dates=False)

##inserting data into mysql
deeds_file_clean_ins.to_sql('deeds', con=engine, if_exists='append',index=False,chunksize=1)
```
