```python
# Dependencies
import pandas as pd
import sqlalchemy

# Store filepath in a variable
exempt_file = "Real_acct_owner/t_jur_exempt.csv"

# Read our Data file with the pandas library
exempt_file_df = pd.read_csv(exempt_file, encoding="ISO-8859-1")


exempt_file_df = exempt_file_df[["Account", "Tax_dist","Tax_dist_name","Exempt_cat","Exempt_dsc"]]
exempt_file_df

```

```python
##Creating output file with only certain columns
exempt_file_df.to_csv("Real_acct_owner/exempt_file_df.csv", index=False, header=True)
```

```python
## Removing duplicate 

exempt_file_df.drop_duplicates(subset='Account', inplace=True)
exempt_file_df.to_csv("Real_acct_owner/exempt_file_df_dup.csv", index=False, header=True)
```

```python
## connection for Mysql
engine = sqlalchemy.create_engine('mysql+pymysql://root:root@localhost:3306/hcad')

## reading file to insert into mysql
owner_data_ins =  pd.read_csv('Real_acct_owner/exempt_file_df_dup.csv', encoding='latin-1',usecols=['Account', 'Tax_dist','Tax_dist_name','Exempt_cat','Exempt_dsc'])

##inserting data into mysql
owner_data_ins.to_sql('exempt_data', con=engine, if_exists='append',index=False,chunksize=1)
```
